```python
from datasets import load_from_disk
import matplotlib.pyplot as plt
# Load the saved dataset using the correct relative path
dataset = load_from_disk("../../data/processed/hf_dog_breeds")

# Print the basic structure
print(dataset)

print(dataset[0])

plt.imshow(dataset[0]["image"])




```

```python
import pandas as pd

# Convert the relevant dataset columns to a Pandas DataFrame
df = dataset.select_columns(["unified_label_id", "unified_label_text"]).to_pandas()

# Get the unique mappings and sort them by ID
breeds_df = df.drop_duplicates().sort_values('unified_label_id').reset_index(drop=True)

# Export to CSV
csv_path = "breed_mapping.csv"
breeds_df.to_csv(csv_path, index=False)

print(f"Exported {len(breeds_df)} breeds to {csv_path}")
# Display the first few rows
breeds_df.head()
```
